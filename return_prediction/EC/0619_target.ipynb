### 該品類(7,14,30,60,180天銷量,比例) 最大值 最小值 中位數 平均數

```python
orderdf.head(3)
```

```python
orderdf.sort_values(by='business_hour_guid').head(10)
```

```python
codetable=orderdf.drop_duplicates('code')[['code']]
codetable.head(3)
```

```python
for i in codetable.code.unique():
    code=i
    temptable=orderdf[orderdf.code==code]
    for j in [7,14,30,60,180]:
        subtable=temptable[temptable.day_diff<j].groupby('business_hour_guid')['subtotal'].sum().to_frame().reset_index()
        codetable.loc[codetable.code==code,f'code_mean_{j}days']=subtable['subtotal'].mean()
        codetable.loc[codetable.code==code,f'code_max_{j}days']=subtable['subtotal'].max()
        codetable.loc[codetable.code==code,f'code_min_{j}days']=subtable['subtotal'].min()
        codetable.loc[codetable.code==code,f'code_median_{j}days']=subtable['subtotal'].median()
        
        
```

```python
codetable=codetable.fillna(0).reset_index(drop=True)
```

```python
#過去該品類消費記錄
codehistory=orderDataframe[['code','business_hour_guid','buying_date','subtotal']]
codehistory.head(3)
```

```python
totaldatetable=pd.DataFrame(columns=['code','date'])
for i in codehistory.code.unique():
    datetable=pd.DataFrame()
    datetable['date']=pd.date_range('1/1/2014','4/30/2019')
    datetable['code']=i
    totaldatetable=pd.concat([totaldatetable,datetable],axis=0)
```

```python
totaldatetable=totaldatetable.sort_values(by=['code','date'])
totaldatetable=totaldatetable.rename(columns={'date':'buying_date'})
print(totaldatetable.shape)
totaldatetable.head(3)
```

```python
codehistory=codehistory.sort_values(by=['buying_date','code']).reset_index(drop=True)
print(codehistory.shape)
codehistory.head(3)
```

```python
codebyday=codehistory.groupby(['code','business_hour_guid','buying_date'])['subtotal'].sum().reset_index().sort_values(by=['buying_date','code']).reset_index(drop=True)
```

```python
codebyday.head(10)
```

```python
#找出每天每個code的最大最小值
maxtable=codebyday.groupby(['code','buying_date'])['subtotal'].max().to_frame().reset_index().rename(columns={'subtotal':'subtotal_max'})
mintable=codebyday.groupby(['code','buying_date'])['subtotal'].min().to_frame().reset_index().rename(columns={'subtotal':'subtotal_min'})
mediantable=codebyday.groupby(['code','buying_date'])['subtotal'].median().to_frame().reset_index().rename(columns={'subtotal':'subtotal_median'})
meantable=codebyday.groupby(['code','buying_date'])['subtotal'].mean().to_frame().reset_index().rename(columns={'subtotal':'subtotal_mean'})
```

```python
sumtable=codebyday.groupby(['code','buying_date'])['subtotal'].sum().to_frame().reset_index().rename(columns={'subtotal':'subtotal_sum'})
```

```python
sumtable
```

```python
maxtable=pd.merge(totaldatetable,maxtable,how='left').fillna(0)
mintable=pd.merge(totaldatetable,mintable,how='left').fillna(0)
mediantable=pd.merge(totaldatetable,mediantable,how='left').fillna(0)
meantable=pd.merge(totaldatetable,meantable,how='left').fillna(0)
meantable.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
##max
maxfinaldf=pd.DataFrame(columns=['code', 'buying_date','code_before7days_max',
       'code_before14days_max', 'code_before30days_max',
       'code_before60days_max', 'code_before180days_max'])
lista=[]
for k in codehistory.code.unique():
    lista.append(k)
    maxdf=maxtable[maxtable.code==k].reset_index(drop=True)
    if(len(lista)%50==0):
        print(f'{len(lista)}/ {len( codehistory.code.unique())}')
    for i in [7,14,30,60,180]:
        maxdf[f'code_before{i}days_max']=maxdf['subtotal_max'].rolling(7).max()
        maxdf=maxdf.fillna(0)
    maxfinaldf=pd.concat([maxfinaldf,maxdf],axis=0).reset_index(drop=True)
maxfinaldf=maxfinaldf.drop(columns='subtotal_max')
```

```python
##mean
meanfinaldf=pd.DataFrame(columns=['code', 'buying_date','code_before7days_mean',
       'code_before14days_mean', 'code_before30days_mean',
       'code_before60days_mean', 'code_before180days_mean'])
lista=[]
for k in codehistory.code.unique():
    lista.append(k)
    meandf=meantable[meantable.code==k].reset_index(drop=True)
    if(len(lista)%50==0):
        print(f'{len(lista)}/ {len( codehistory.code.unique())}')
    for i in [7,14,30,60,180]:
        meandf[f'code_before{i}days_mean']=meandf['subtotal_mean'].rolling(7).mean()
        meandf=meandf.fillna(0)
    meanfinaldf=pd.concat([meanfinaldf,meandf],axis=0).reset_index(drop=True)
meanfinaldf=meanfinaldf.drop(columns='subtotal_mean')
```

```python
##min
minfinaldf=pd.DataFrame(columns=['code', 'buying_date','code_before7days_min',
       'code_before14days_min', 'code_before30days_min',
       'code_before60days_min', 'code_before180days_min'])
lista=[]
for k in codehistory.code.unique():
    lista.append(k)
    mindf=mintable[mintable.code==k].reset_index(drop=True)
    if(len(lista)%50==0):
        print(f'{len(lista)}/ {len( codehistory.code.unique())}')
    for i in [7,14,30,60,180]:
        mindf[f'code_before{i}days_min']=mindf['subtotal_min'].rolling(7).min()
        mindf=mindf.fillna(0)
    minfinaldf=pd.concat([minfinaldf,mindf],axis=0).reset_index(drop=True)
minfinaldf=minfinaldf.drop(columns='subtotal_min')
```

```python
##median
medianfinaldf=pd.DataFrame(columns=['code', 'buying_date', 'code_before7days_median',
       'code_before14days_median', 'code_before30days_median',
       'code_before60days_median', 'code_before180days_median'])
lista=[]
for k in codehistory.code.unique():
    lista.append(k)
    mediandf=mediantable[mediantable.code==k].reset_index(drop=True)
    if(len(lista)%50==0):
        print(f'{len(lista)}/ {len( codehistory.code.unique())}')
    for i in [7,14,30,60,180]:
        mediandf[f'code_before{i}days_median']=mediandf['subtotal_median'].rolling(7).median()
        mediandf=mediandf.fillna(0)
    medianfinaldf=pd.concat([medianfinaldf,mediandf],axis=0).reset_index(drop=True)
medianfinaldf=medianfinaldf.drop(columns='subtotal_median')
```

```python
assert medianfinaldf.shape==meanfinaldf.shape
assert medianfinaldf.shape==minfinaldf.shape
assert medianfinaldf.shape==maxfinaldf.shape
```

```python
totalfinaldf=pd.merge(meanfinaldf,medianfinaldf,on=['buying_date','code'])
totalfinaldf=pd.merge(totalfinaldf,maxfinaldf,on=['buying_date','code'])
totalfinaldf=pd.merge(totalfinaldf,minfinaldf,on=['buying_date','code'])
totalfinaldf.head(3)
```
