```python
import pandas as pd
import numpy as np
```

```python
deal=pd.read_csv('/Users/leetsaiying/Desktop/17life/0605/deal(501mb;546Krows)_0605v1.csv')
```

```python
#欄位有問題 處理欄位名字
deal=deal.reset_index()
deal.columns=(['business_hour_guid', 'seller_GUID', 'business_hour_order_time_s',
       'business_hour_order_time_e', 'business_hour_deliver_time_s',
       'business_hour_deliver_time_e', 'pic_alt','tzip_code', 'city_name',
       'district_name', 'item_name', 'item_orig_price', 'item_price',
       'item_default_daily_amount', 'event_name', 'event_title',
       'product_spec', 'shopping_cart', 'combo_pack_count',
       'quantity_multiplier', 'city_list', 'deal_acc_business_group_id',
       'acc_business_group_name', 'deal_type', 'is_travel_deal',
       'installment_3months', 'installment_6months', 'installment_12months',
       'deny_installment', 'shippingdate', 'is_long_contract',
       'develope_sales_id', 'operation_sales_id', 'slug', 'ordered_quantity',
       'ordered_total', 'category_list', 'freight_amount', 'enable_isp',
       'discount_price', 'is_wms', 'ancestor_business_hour_guid', 'code_name',
       'code'])
```

```python
deal.head(3)
```

```python
#選取所需欄位作新表
dealtemp=deal[['business_hour_guid','seller_GUID','item_default_daily_amount','quantity_multiplier','ordered_quantity']]
```

```python
#檢視一下資料狀況
dealtemp.head(50)
```

```python
#檢查有無空值
deal['item_default_daily_amount'].isna().sum()
```

```python
#groupby檔期id並存入dealgroup
dealgroup=dealtemp.groupby('business_hour_guid')
```

```python
#先以一個id作測試
dealgroup.get_group('795DE868-0304-4ED9-A5EA-A8C995071628')['item_default_daily_amount'].sum()
```

```python
#測試時間用
%%time
sumofamount=[]
for i in dealtemp['business_hour_guid']:
    a=dealgroup.get_group(i)['item_default_daily_amount'].sum()
    sumofamount.append(a)
    
    
    
```

```python
#做全部檔期的數量加總
sumofamount=[]
for i in dealtemp['business_hour_guid']:
    a=dealgroup.get_group(i)['item_default_daily_amount'].sum()
    sumofamount.append(a)
```

```python
type(sumofamount)
```

```python
#轉型成dataframe
sumofamount_df=pd.DataFrame(sumofamount)
```

```python
#檢查是否為dataframe
type(sumofamount_df)
```

```python
type(dealgroup)
```

```python
#將dealgroup轉型為dataframe
dealgroup_df=pd.DataFrame(dealgroup)
```

```python
#欄位改名
dealgroup_df.columns=['guidid','no']
```

```python
#將組別與總數量兩個表合併 為totalofguid
totalofguid=pd.concat([dealgroup_df,sumofamount_df],axis=1)
```
