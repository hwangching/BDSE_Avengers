```python
#order left join deal
dftemp=pd.merge(order,deal,on="business_hour_guid",how='left')
```

```python
dftemp.columns
```

```python
#選取所需欄位 製作新表df
df=dftemp[['GUID','seller_GUID_x','subtotal','business_hour_guid','item_price','quantity_multiplier']]
```

```python
#檢視
df.head()
```

```python
#quantity_multiplier的空值數
df['quantity_multiplier'].isna().sum()
```

```python
#以1取代空值
df['quantity_multiplier']=df['quantity_multiplier'].fillna(1)
```

```python
#檢查空值數是否為0
df['quantity_multiplier'].isna().sum()
```

```python
#計算商品單價 並 併入df表裡面 （商品價格/包裝份數）
df['unit_price']=df['item_price']/df['quantity_multiplier']
df['unit_price']
```

```python
df.head(2)
```

```python
#計算各筆的銷售數量（總額/商品單價）
df['unit_amount']=df['subtotal']/df['unit_price']
df['unit_amount']
```

```python
#檢視
df.head(2)
```

```python
sumofamount=df.groupby('business_hour_guid')["unit_amount"].sum()
```
