```python
order=pd.read_csv("order.csv")
```

```python
order.rename(columns={'Unnamed: 4':'is_returned'}, inplace=True)
```

```python
order_returned=order.groupby('user_id')
order_returned.head(3)
```

```python
order_returned.size()
```

```python
order_change=order.sort_values('user_id')
```

```python
order_change.head(5)
```

```python
df =pd.DataFrame()
df['totalamount']=order_change.groupby("user_id")["subtotal"].sum()
```

```python
dfnum=pd.DataFrame()
dfnum['totalnum']=order_change.groupby("user_id")["subtotal"].count()
```

```python
new=pd.merge(df,dfnum,on='user_id')
new['avg_amount']=new['totalamount']/new['totalnum']
```
