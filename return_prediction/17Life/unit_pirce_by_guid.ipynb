```python
#發現“包裝份數”有nan值
deal['quantity_multiplier'].head(10)
```

```python
#nan值用1取代
def trans(x):
    if x>0: 
        return x
    else:
        return 1
deal['quantity_multiplier']=deal['quantity_multiplier'].apply(trans)
```

```python
deal['unit_price']=deal['item_price']/deal['quantity_multiplier']
deal[['business_hour_guid','seller_GUID','unit_price']]
```
