### 進行tf-idf + 自動分群

```python
# 建立 tf-idf 模型
n_features = 800
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                max_features=n_features,
                                stop_words='english',
                                max_df = 0.85,
                                min_df = 3)
```

```python
# 餵進資料
tf = tf_vectorizer.fit_transform(df.content_cutted)
```

```python
# 觀看模型的產出
tf_feature_names = tf_vectorizer.get_feature_names()
print(len(tf_feature_names))
tf_feature_names
```

```python
# 開始進行分群
n_topics = range(5,105, 5)
perplexityLst = [1.0]*len(n_topics)
lda_models = []
for idx, n_topic in enumerate(n_topics):
    lda = LatentDirichletAllocation(n_components=n_topic,
                                    max_iter=50,
                                    learning_method='online',
                                    evaluate_every=200,
                                    verbose=0)

    lda.fit(tf)
    perplexityLst[idx] = lda.perplexity(tf)    # 計算模型的困惑度，分數越低越好
    lda_models.append(lda)
    print ("分成 {} 群".format(n_topics[idx]))
    print ("Perplexity Score %0.3f" % perplexityLst[idx])
```

```python
#打印最佳模型
best_index = perplexityLst.index(min(perplexityLst))
best_n_topic = n_topics[best_index]
best_model = lda_models[best_index]
print ("Best # of Topic: ", best_n_topic)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("第 {} 群的top 10 :".format(topic_idx))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
```

```python
n_top_words = 10
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```
