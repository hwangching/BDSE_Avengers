```python
# -*- coding: utf-8 -*-
import os
import pandas as pd
import numpy as np
import join
import jieba
import jieba.analyse
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import time

print("【訊息】目前的工作目錄:",os.getcwd())
```

```python
datafile=pd.read_excel(r'D:\Mypython\專題\deal.xlsx', sep=',', index_col=0, header=0)
datafile.head()
```

```python
def word_cut(mytext):
    return " ".join(jieba.cut(mytext))
```

```python
datafile["content_cutted"] = datafile.品名加子分類名.apply(word_cut)
```

```python
datafile["content_cutted"]
```

```python
n_features = 200
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                max_features=n_features,
                                stop_words='english',
                                max_df = 0.95,
                                min_df = 10)
tf = tf_vectorizer.fit_transform(datafile.content_cutted)
```

```python
tf_feature_names = tf_vectorizer.get_feature_names()
print(len(tf_feature_names))
tf_feature_names
```

```python
n_topics = 10
lda = LatentDirichletAllocation(n_components=n_topics, max_iter=50,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
```

```python
n_topics = range(10, 65, 5)
perplexityLst = [1.0]*len(n_topics)
lda_models = []
for idx, n_topic in enumerate(n_topics):
    lda = LatentDirichletAllocation(n_components=n_topic,
                                    max_iter=50,
                                    learning_method='online',
                                    evaluate_every=200,
                                    verbose=0)

    lda.fit(tf)
    perplexityLst[idx] = lda.perplexity(tf)
    lda_models.append(lda)
    print ("# of Topic: {},".format(n_topics[idx]))
    print ("Perplexity Score %0.3f" % perplexityLst[idx])

#打印最佳模型
best_index = perplexityLst.index(min(perplexityLst))
best_n_topic = n_topics[best_index]
best_model = lda_models[best_index]
print ("Best # of Topic: ", best_n_topic)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
```

```python
n_top_words = 25
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```
